```python
import numpy as np

def gred(x,y):
    wc=bc=0
    it = 1000
    n = len(x)
    lr = 0.000
    for i in range(it):
        yp = wc*x + bc
        cost = (1/n) * sum([val**2 for val in (y-yp)])
        
        wd = (2/n)*sum(x*(y-yp))
        bd = (2/n)*sum(y-yp)
        wc = wc - lr*wd
        bc = bc - lr*bd
        print ("m-{}, b-{}, cost-{}, it-{}".format(wc,bc,cost,i))        
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,8,11,13])

gred(x,y)
```
